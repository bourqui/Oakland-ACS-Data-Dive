## 2. Build the SQL to get that data out of BigQuery

Not pictured is a lot of poking around in BigQuery's SQL client itself, looking for the right ACS tables and columns...



... but eventually I got where I was going. I found that I wanted to use the ACS `place_YEAR_1yr` tables to get a sense of the yearly survey data and how it was evolving year-over-year, which for my purposes I find more interesting than the every-ten-years census. (The screenshot above is of the `state_YEAR_1yr` table -- ACS divides up the data by many different chunks of geographic units, helpfully -- but I mostly used the `place` tables to get the city of Oakland.)

````python
interesting_tables=[]

#We still have our start_year, end_year, and range_years we can use:
for year in list(range(2007,2019)):
    table_to_add=f"place_{year}_1yr"
    interesting_tables.append(table_to_add)
    
print(interesting_tables)
````

At this point I began to think: I definitely don't need 252 columns of information, and even if I trimmed out the five that the others don't have to make a nice SQL UNION of 247 columns, better to just get a targeted query of what I want now and then query again if I have to. The ACS data is only one row per table, so it's not like the queries are expensive or hard.

With that in mind I'm shifting my strategy again to find the columns I most want to focus on up-front, which is probably where I should have started. I'll look through the schema of one of the year tables in the BigQuery UI and focus on what feels most interesting:



Looking through this, there's a number of columns I'd like to explore:

- same-sex households
- Car-free households ('no_cars')
- High income (200,000+)

Now I'll need to add these to a list, get them into a SQL query, and then find the UNION of each of the years' data.

````python
cols=['median_income', 
      'income_per_capita',
      'families_with_young_children',
      'million_dollar_housing_units',
      'bachelors_degree_or_higher_25_64',
      'different_house_year_ago_different_city',
      'households_public_asst_or_food_stamps',
      'high_school_including_ged',
      'graduate_professional_degree',
      'male_male_households',
      'female_female_households',
      'no_cars',
      'income_200000_or_more',
      'households' # I want to get all households N to do percentages
     ]

#Now to collate them into a SQL query that can be duplicated.
fieldlist=str(cols)
chars_to_strip="[]\'"
for char in chars_to_strip:
    fieldlist=fieldlist.replace(char,'') #strip out list characters
````

I use a for loop to get the SQL I want, since I'm going to UNION a number of tables each representing one year of ACS data for the city of Oakland.

I'll also need to add a Year column to the query since the tables' year is specified in each of the table names, not in its own column. I add that below.

````python
# Time to generate the SQL query I want.

# Set variables for query
oakland='0653000' #City of Oakland

def make_union_query(some_fields,some_range,some_place):
    query_string=""
    for year in list(some_range):
        sql=f"SELECT \'{year}\' AS Year, {some_fields} FROM `bigquery-public-data.census_bureau_acs.place_{year}_1yr` WHERE geo_id='{some_place}'\n"
        query_string+=sql
        if year==list(some_range)[-1]: # if it's the final year in the list ... 
            pass  # ... don't add a UNION on the very end of the query
        else:
            union="UNION ALL\n"
            query_string+=union
    return query_string

# I do a range from 2007 to 2018, but range syntax means entering that as 2019
query=make_union_query(fieldlist,range(2007,2019),oakland)
print(query)
````

## 3. Query the data out of BigQuery and into a Pandas dataframe

Now let's feed it back into BigQuery:

````python
bqclient = bigquery.Client()

dataframe = (
    bqclient.query(query)
    .result()
    .to_dataframe(
        create_bqstorage_client=True,
    )
)
dataframe=dataframe.sort_values(by='Year')
````

## 4. Analyze the data

Time to explore!

I'd like to start by seeing how same-sex households are evolving as a percentage of all households over time.

````python
# Get households as percentage, not as absolute N; I want to see %
x=dataframe['Year'].astype(int)
y=dataframe['male_male_households']/dataframe['households']

# Create figure and axes
fig=plt.figure(figsize=[10,5],dpi=300)
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

# Use numpy to calculate slope and y-intercept of best fit line
m,b=np.polyfit(x,y,1) 

# Now plot the values and best fit line
ax1.plot(x,y)
ax1.plot(x,m*x+b,label='Trend')
ax1.set_title('Male same-sex households')

#Now let's do the same for female-female households
y=dataframe['female_female_households']/dataframe['households']
m,b=np.polyfit(x,y,1) #recalculate best fit line
ax2.plot(x,y,label='Self-reported households as a \npercentage of Oakland households')
ax2.plot(x,m*x+b,label='Trend')
ax2.set_title('Female same-sex households')

# Set formatting and y-range on both axes
axes=(ax1,ax2)
for ax in axes:
    ax.set_ylim(bottom=0,top=.02)
    vals = ax.get_yticks()
    ax.set_yticklabels(['{:,.1%}'.format(x) for x in vals])
    for location in ['left', 'right', 'bottom', 'top']:
        ax.spines[location].set_visible(False)
    ax.tick_params(bottom=False, left=False)

plt.legend()
plt.show()
````

Here we can start to see some interesting trends despite the noisiness of the survey data. Despite a wide variance, we can see a decline in same-sex households, both male and female, with an especially clear decline in female-female households in ten years from ~1% of the population to about 0.5%. 

Now I'd like to explore income. I'll be looking at `median_income` (which is the citywide median household income) and `income_per_capita` (which is mean income for every person.)

````python
# Our x-axis is already set as Year correctly, so:

# # Draw plot
fig=plt.figure(figsize=[10,5],dpi=300)
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

# # Median income
y1=dataframe['median_income']
ax1.plot(x,y1)
ax1.set_title('Oakland median household income')

# # Income per capita
y2=dataframe['income_per_capita']
ax2.plot(x,y2)
ax2.set_title('Oakland income per capita')

# Set formatting and y-range on both axes
axes=(ax1,ax2)
max_val=max(max(y1),max(y2)) # Get maximum y-value of both series
for ax in axes:
    ax.set_ylim(bottom=0,top=max_val+10000)
    for location in ['right','top']:
        ax.spines[location].set_visible(False)
    ax.tick_params(bottom=False, left=False)
    
plt.show()
````

These numbers are pretty stunning. However, I'm realizing I want to see it compared to California as a whole. Fortunately, I can requery the data now that we have the field list already set up to ask BigQuery, and I'll just replace the `geoid` with California's code (`06`). 

In order to do so I'll have to slightly tweak the query function I built to reference the state tables rather than the place tables.

````python
# Redefine the query-writing function to reference the correct tables
def make_state_query(some_fields,some_range,some_place):
    query_string=""
    for year in list(some_range):
        sql=f"SELECT \'{year}\' AS Year, {some_fields} FROM `bigquery-public-data.census_bureau_acs.state_{year}_1yr` WHERE geo_id='{some_place}'\n"
        query_string+=sql
        if year==list(some_range)[-1]:
            pass #don't add a UNION on the very end of the query
        else:
            union="UNION ALL\n"
            query_string+=union
    return query_string

# This makes our query with our list of fields we set before.
# I use CA's geoid of '06'
ca_query=make_state_query(fieldlist,range(2007,2019), '06') 

# And this gets our dataframe
cadf = (
    bqclient.query(ca_query)
    .result()
    .to_dataframe(
        create_bqstorage_client=True,
    )
).sort_values(by='Year')

cadf # Let's confirm it's there
````

Great. Now let's add the California values to the chart above.

````python
# # Draw plot
fig=plt.figure(figsize=[10,5],dpi=300)
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

# Median income
y1=dataframe['median_income']
ax1.plot(x,y1,label='Oakland')
y1_ca=cadf['median_income']
ax1.plot(x,y1_ca,label='California')
ax1.set_title('Oakland median household income')

# Income per capita
y2=dataframe['income_per_capita']
ax2.plot(x,y2,label='Oakland')
y2_ca=cadf['income_per_capita']
ax2.plot(x,y2_ca,label='California')
ax2.set_title('Oakland income per capita')

# Set formatting and y-range on both axes
axes=(ax1,ax2)
max_val=max(max(y1),max(y2)) # Get maximum y-value of both series
for ax in axes:
    ax.set_ylim(bottom=0,top=max_val+10000)
    for location in ['right','top']:
        ax.spines[location].set_visible(False)
    ax.tick_params(bottom=False, left=False)
plt.legend()
plt.show()
````

From here we can see that Oakland has caught up with California's median household income after historically lagging. Meanwhile, with regards to income per capita, Oakland is pulling away from the California income per capita average.

A few reflections and hypotheses on what we may be seeing here:
- Median household income may be rising as Oakland gentrifies and more wealthy families are pushed out of San Francisco as the cost of living in the Bay Area, and particularly in job centers like San Francisco and Palo Alto, have risen incredibly over the past ten years.
- Oakland has traditionally been a more underinvested community, and even with the median household income rising, $75k doesn't go as far in Oakland as it does in the rest of California.
- Perhaps the reason mean per capita income is rising more slowly than median household income may be to do with the fact that households might be greater carriers of wealth than individuals; I'd be curious to know how the preponderence of homeless, students, and single people push down the median household income vs. other cities.
